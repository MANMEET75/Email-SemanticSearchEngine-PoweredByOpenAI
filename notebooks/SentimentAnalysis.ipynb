## Testing with MultinormalNB

```python
#Loading the Dataset
import pandas as pd
data = pd.read_csv('data/data1.csv')
data.dropna(axis=0,inplace=True)
#Pre-Prcoessing and Bag of Word Vectorization using Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(data['body_new'])
#Splitting the data into trainig and testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(text_counts, data['Sentiment'], test_size=0.25, random_state=5)
#Training the model
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, Y_train)
#Caluclating the accuracy score of the model
from sklearn import metrics
predicted = classifier.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, Y_test)
print("Accuracuy Score: ",accuracy_score)
```

## Testing with KNeighbors Classifier

```python
#Loading the Dataset
import pandas as pd
data = pd.read_csv('data/data1.csv')
data.dropna(axis=0,inplace=True)
#Pre-Prcoessing and Bag of Word Vectorization using Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(data['body_new'])
#Splitting the data into trainig and testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(text_counts, data['Sentiment'], test_size=0.25, random_state=5)
#Training the model
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
classifier.fit(X_train,Y_train)
#Caluclating the accuracy score of the model
from sklearn import metrics
predicted = classifier.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, Y_test)
print("Accuracuy Score: ",accuracy_score)
```

## TFIDF

```python
# Loading the Dataset
import pandas as pd
data = pd.read_csv('data/data1.csv')
data.dropna(axis=0, inplace=True)

# Pre-Processing and TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import RegexpTokenizer

token = RegexpTokenizer(r'[a-zA-Z0-9]+')
tfidf = TfidfVectorizer(stop_words='english', tokenizer=token.tokenize)
text_tfidf = tfidf.fit_transform(data['body_new'])

# Splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(text_tfidf, data['Sentiment'], test_size=0.25, random_state=5)

# Training the model
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
classifier.fit(X_train, Y_train)

# Calculating the accuracy score of the model
from sklearn import metrics
predicted = classifier.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, Y_test)
print("Accuracy Score: ", accuracy_score)
```

## Cleaning the text

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

nltk.download('stopwords')

corpus = []
for i in range(len(dataset)):
    try:
        body_new = str(dataset['body_new'].iloc[i])
        body_new = re.sub('[^a-zA-Z]', ' ', body_new)
        body_new = body_new.lower()
        body_new = body_new.split()
        ps = PorterStemmer()
        all_stopwords = stopwords.words('english')
        all_stopwords.remove('not')
        body_new = [ps.stem(word) for word in body_new if word not in all_stopwords]
        body_new = ' '.join(body_new)
        corpus.append(body_new)
    except KeyError:
        print(f"KeyError: 'body_new' column does not exist in the dataset.")


```

## Creating the model using Word2Vec

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
len(corpus)
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent)) 
```
